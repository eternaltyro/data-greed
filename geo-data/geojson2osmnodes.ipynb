```python
from osmapi import OsmApi
import geojson

#api_endpoint = "api06.dev.openstreetmap.org"
api_endpoint = "api.openstreetmap.org"
password_file = "/tmp/password.osm"
datafile = "libraries_new.geojson"
handle = open(datafile)
geo_data = geojson.load(handle)
handle.close()
```

```python
# Handle GeoJSON
features = geo_data.get("features")
def magic():
    for amenity in features:
        coords, tags = amenity["geometry"]["coordinates"], amenity["properties"]
        del tags["marker-color"]
        del tags["marker-symbol"]
        yield { u'lat': coords[1], u'lon': coords[0], u'tag': tags }
```

```python
# Handle Connections
connection_handle = OsmApi(passwordfile=password_file, api=api_endpoint)
changesetID = connection_handle.ChangesetCreate({u'comment': u'Public libraries (mostly government operated) in Chennai surveyed manually with GPS', u'source': u'GPS and Survey'})
library_nodes = []
for node in magic():
    newNode = connection_handle.NodeCreate(node)
    library_nodes.append(newNode)
connection_handle.flush()
connection_handle.ChangesetClose()
```
